```python
import sidrapy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Dados da CNT - Contas nacionais trimestrais
#https://apisidra.ibge.gov.br/values/t/1621/n1/all/v/all/p/all/c11255/90707,93406/d/v584%201
df_cnt = sidrapy.get_table(
    table_code='1621',
    territorial_level = "1",
    ibge_territorial_code = "all",
    variable="584",
    period = "all",
    classifications= {'11255':'90707,93406'},
    header='n')
```

```python
#--> PIB
df_pib = df_cnt.loc[df_cnt['D4N']=='PIB a preços de mercado'].reset_index(drop=True)
pib = df_pib.loc[:,('D2C','V')]
pib['D2C'] = pd.date_range('1996', periods=len(df_pib), freq='QS')
pib.rename(columns={'D2C':'data','V':'valor'}, inplace=True)
```

```python
#--> Formação bruta de capital fixo
df_fbcf = df_cnt.loc[df_cnt['D4N']=='Formação bruta de capital fixo'].reset_index(drop=True)
fbcf = df_fbcf.loc[:,('D2C','V')]
fbcf['D2C'] = pd.date_range('1996', periods=len(df_fbcf), freq='QS')
fbcf.rename(columns={'D2C':'data','V':'valor'}, inplace=True)
```

```python
#--> Taxa de investimento
tx_inv = df_inv.loc[:,('D2C','V')]
tx_inv['D2C'] = pd.date_range('1996', '2024-01-01', freq='QS')
tx_inv.rename(columns={'D2C':'data','V':'valor'}, inplace=True)
```

```python
#seleciona os dados numa data especificada
def get_data(dado,ini,fim):
    index_ini = dado.loc[dado['data'] == ini].index[0]
    index_fim = dado.loc[dado['data'] == fim].index[0]

    dado = dado.iloc[index_ini:index_fim].reset_index(drop=True)

    return dado

#transforma o dataframe em time series devolvendo apenas os dados
def get_ts(dado):
    dado.index = dado['data']
    dado = dado['valor']

    return dado
```
